```python
import random
import pandas as pd

def select_random_text(csv_file):
    # Read the CSV file into a pandas DataFrame
    df = pd.read_csv(csv_file)

    # Select a random index
    random_index = random.randint(0, len(df) - 1)

    # Get the random text from the 'text' column
    random_text = df.loc[random_index, 'text']

    return random_text

# Example usage:
random_text = select_random_text('/content/rnn1.csv')
print(random_text)
text1=random_text
```

```python
out = pipe(text1)
result = out[0]  # Assuming you want the first result if multiple are returned
sentiment = result["label"]
score = round(result["score"], 2)
```

```python
print(f"Sentiment: {sentiment}")
print(score)
```
